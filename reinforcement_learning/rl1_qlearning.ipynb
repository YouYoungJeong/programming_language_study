# 강화학습(Reinforcement Learning)
### Q-learning의 구조를 이해하기
- 벨만 방정식(Bellman Equation) 기반의 근사학습
- Q-learning에서 에이전트는 가장 Q값이 높은 행동을 선택(greedy한 행동이다.)
---
현재 위치(state)에서 어떤 행동(action)을 취할 것인가? 왼쪽, 오른쪽

```python
import numpy as np
import random

# 상태 공간 : 에이전트가 있을 수 있는 위치
state_space = [0, 1, 2, 3, 4]

# 행동 공간 : [-1 왼쪽, 1 오른쪽] 이동을 의미
action_space = [-1, 1]

# Q-table : (행 : 상태, 열 : 행동)
# Q[state][action] <= 특정 상태에서 특정 행동을 했을 때의 가치를 표현하는 방법

# 초기치
Q = np.zeros((len(state_space), len(action_space)))
print(Q)
```

```python
# Hyper 파라미터 설정하기
alpha = 0.1   # 학습률(새롭게 배운값을 기존 Q값에 얼마나 반영하지 결정)
gamma = 0.9   # Discount Factor(할인율, γ) - 미래 보상을 얼마나 중요하게 볼 것인지를 결정
epsilon = 1.0 # 탐험 확률값(엡실론, ε)
epsilon_decay = 0.99  # epsilon 감소율
epsilon_min = 0.1     # 탐험 최소 확률(최소 10%만큼은 탐험을 해라)
episodes = 500 # 전체 학습(에피소드는 한번의 도전) 횟수

# 보상 함수 생성하기 -> state가 4이면 목표에 도달
def get_reward(state):
  return 10 if state == 4 else 0
```

```python
from __future__ import barry_as_FLUFL
# 학습 시작하기 : episode는 하나의 학습 시도
for episod in range(episodes):
  state = 0 # 매 에피소드마다 0번 위치에서 시작

  # 에피소드가 지속적으로 반복되면 강제로 멈춰줘야함
  # 한 에피소드 안에서 행동(action)은 20번으로 제한함
  for step in range(20): # step:한번의 이동
    # --------- 행동 선택 ---------
    # 탐험(Exploration) - 랜덤함
    if random.random() < epsilon: # 에이전트의 랜덤값(0< 임의의 실수값 <1)보다 작으면 탐험
      action_index = random.randint(0, 1) # 0 또는 1을 랜덤으로 선택

    # 이용(Exploitation) - greedy(탐욕)적 행동을 함.
    else:
      action_index = np.argmax(Q[state]) # greedy한 action : Q값이 큰쪽의 행동을 취함.

    # --------- 실제 행동 값 주기 ---------
    action = action_space[action_index] # 실제 행동값(-1 or 1)로 변환 시킴.
    # print('action :',action)

    # --------- 다음 상태 계산 ---------
    # ex) state가 2 이고 action이  1인 경우 => 다음 상태는 3
    # ex) state가 2 이고 action이 -1인 경우 => 다음 상태는 1
    next_state = state + action

    # --------- 상태 공간 유지 ---------
    # 현재 상태 공간이 0보다 작으면 안되고 4보다 크면 현재 상태 계속 유지.
    if next_state < 0 or next_state > 4:
        next_state = state

    # --------- 보상 받기 ---------
    reward = get_reward(next_state)

    # --------- Q값 갱신 ---------
    old_q = Q[state][action_index]   # 현재 상태
    next_max = np.max(Q[next_state]) # 다음 상태에서 가장 좋은 행동을 했을 때 기대되는 가치

    # --------- Q-learning 갱신 식 ---------
    # 벨만 방정식 응용 : off-policy 방식의 수식
    Q[state][action_index] = old_q + alpha * (reward + gamma * next_max - old_q)

    # --------- 상태 이동 --------
    # 다음 상태를 현재 상태로 변경
    state = next_state

    # --------- 보상을 보고 판단 --------
    if reward == 10:
      break

  # --------- 다음 에피소드로 이동 --------
  # epsilon 감소 - 탐험 감소 이용 증가
  epsilon = max(epsilon_min, epsilon * epsilon_decay)
  # print("epsilon :",epsilon)

print(Q)

```
