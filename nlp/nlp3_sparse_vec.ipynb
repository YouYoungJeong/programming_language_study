# 희소벡터 표현 방식
## CountVectorizer
- 단어 등장 횟수 기반 - 정수로 표현
- 단어 가 몇번 나왔는가가 중요ㄴ

##TfidfVectorizer
- 문서에서 자주 나타나는 단어에 가중치를 높게 주되, 모든(여러) 문서에서 자주 등장하는 단어에는 패널티를 부여 - 실수로 표현
- 그 단어가 문서에서 얼마나 중요한가를 따짐
### ∴ CountVectorizer, TfidfVectorizer 는 '문장 안에 어떤 단어가 들어 있는가?'를 바꾸는 방식
    -> 키워드가 정확히 맞아야 하는 검색에 효과적

### Word2Vec
- 그 단어가 어떤 의미적 위치에 있는가를 따짐
- '단어가 어떤 의미적 위치에 있는가'를 숫지로 바꾸는 방식
- 의미가 비슷한 단어까지 찾을때 효과적


```python
# CountVectorizer 연습
print(f"{'-'*20} CountVectorizer {'-'*20}")
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

content = ['How to format my hard disk', 'Hard disk format problems']

count_vec = CountVectorizer(analyzer='word', min_df=1)
print(count_vec)
tran = count_vec.fit_transform(content)
print(tran)
print(count_vec.get_feature_names_out())
print(tran.toarray())
# ['disk' 'format' 'hard' 'how' 'my' 'problems' 'to']
# [[1 1 1 1 1 0 1]
#  [1 1 1 0 0 1 0]]
```

```python
# TfidfVectorizer 연습
print(f"{'-'*20} TfidfVectorizer {'-'*20}")
print(content)
tfidf_vec = TfidfVectorizer(analyzer='word', min_df=1)
# print(tfidf_vec)
tran2 = tfidf_vec.fit_transform(content)
print(tran2)
print(tfidf_vec.get_feature_names_out())
print(tran2.toarray()) # 출현 빈도 확률값 출력
# ['disk' 'format' 'hard' 'how' 'my' 'problems' 'to']
# [[0.33471228 0.33471228 0.33471228 0.47042643 0.47042643 0.
#   0.47042643]
#  [0.44832087 0.44832087 0.44832087 0.         0.         0.63009934
#   0.        ]]
# 0.33471228 : 0번째 문장에서 0번째 단어('disk')는 빈도에 대한 0.3347% 정도의 확률값을 가짐.
```
