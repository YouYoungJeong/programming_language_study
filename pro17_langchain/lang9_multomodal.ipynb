# 1. RAG (검색 증강 생성, Retrieval-Augmented Generation)
AI가 답변하기 전에 외부 데이터베이스나 사내 문서 등에서 관련 정보를 먼저 '검색'하여 가져온 뒤, 이를 바탕으로 정확한 답변을 '생성'하는 기술

핵심: AI가 학습하지 않은 최신 정보나 내부 기밀 문서를 참조하여 환각(할루시네이션)을 줄이고 정확한 사실에 기반한 답변을 하게 해줌

# 2. 에이전틱 AI (Agentic AI)
AI가 수동적인 챗봇을 넘어, 스스로 목표를 설정하고, 판단하며, '행동'하여 업무를 완수하는 자율형 AI 시스템

핵심: 단순히 정보를 요약하는 것을 넘어, "비행기 표를 예매하고 일정에 등록해 줘"라고 하면 예약 시스템에 접속해 결제까지 스스로 처리하는 능력

# 3. AI 에이전트 (AI Agent)
에이전틱 AI를 실제로 구현하기 위해 특정 역할과 목표를 부여받은 독립적인 AI 프로그램(일꾼)

핵심: 기억력, 추론 능력, 도구 사용 능력을 갖추고 있어, 복잡한 작업을 여러 단계로 쪼개서 자율적으로 처리하고 필요하면 다른 AI 에이전트와 협업하기도 함

# 4. Reason + Act (ReAct)
AI가 문제를 해결할 때 추론(Reasoning) 단계와 행동(Acting) 단계를 번갈아 수행하며 정답에 다가가는 프롬프트 및 추론 패턴

핵심: "생각(질문 분석) -> 행동(외부 도구 사용) -> 관찰(결과 확인)"의 루프를 반복하여, 문제의 원인을 파악하고 스스로 오류를 수정하며 고난도 작업을 수행

# 5. MCP (모델 컨텍스트 프로토콜, Model Context Protocol)
다양한 AI 모델이 사내 데이터나 외부 서비스(슬랙, 깃허브, 데이터베이스 등)에 쉽게 연결될 수 있도록 해주는 '표준 연결 규격'

핵심: 마치 다양한 전자기기를 꽂을 수 있는 USB-C 포트처럼, 표준화된 통신 규격을 통해 AI가 개발 환경이나 회사 데이터에 안전하고 매끄럽게 접근할 수 있게 돕는다.

 # 멀티모달 (Multimodal)
 AI의 '눈과 귀' (감각 시스템)

 설명: 텍스트를 넘어 이미지, 음성, 영상 등 다양한 형태의 데이터를 동시에 이해하고 생성하는 기술

 # LangChain (랭체인)
 AI의 '신경망과 도구 가방' (연결 프레임워크)

 설명: AI 모델을 문서(RAG), 데이터베이스, 웹 검색 등 다양한 외부 도구와 사슬처럼 연결해 주는 개발 도구입니다.

 # LangGraph (랭그래프)
 AI의 '뇌 회로' (판단 및 흐름 제어)

 AI가 "생각하고, 행동하고, 검토하는" 복잡한 반복 과정을 그래프 구조로 설계하여 자율적인 업무 수행을 제어하는 도구입니다.


이 세 가지가 합쳐지면 "사람처럼 보고, 판단하고, 도구를 써서 업무를 완수하는 자율형 멀티모달 에이전트"가 완성됩니다.



 [입력: 멀티모달] ➔ [연결: LangChain] ➔ [제어 루프: LangGraph]


 이미지/음성 인식         외부 도구 장착    성공할 때까지 무한 반복



시각적 인지: 멀티모달 기술을 통해 AI가 고장 난 화면 캡처나 영수증 이미지를 눈으로 확인합니다.

도구 사용: LangChain이 제공하는 연결선들을 통해 AI가 필요한 파일 시스템이나 웹 브라우저를 손에 쥡니다.

자율 해결: LangGraph의 설계에 따라 AI는 "수정하기 ➔ 결과 화면 다시 보기(멀티모달) ➔ 실패 시 재시도" 과정을 성공할 때까지 스스로 반복합니다.

```python
from langchain_openai import ChatOpenAI, OpenAIEmbeddings
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough, RunnableLambda
from chromadb import PersistentClient
from langchain_core.messages import HumanMessage
from sentence_transformers import SentenceTransformer
from PIL import Image
import base64
from pathlib import Path
from typing import List, Dict, Any

import os
from dotenv import load_dotenv
load_dotenv()

api_key = os.getenv("OPENAI_API_KEY")
if not api_key:
    raise RuntimeError("OPENAI_API_KEY가 없어요")

llm = ChatOpenAI(model="gpt-4.1-mini", temperature="0.5")
```

```python
# 이미지 , 크로마, 컬렉션 경로 설정하기
IMG_DIR = 'images_db'
CHROMA_DIR = './chroma_image_db'
COLLECTION = "image_docs"

# Hugging Face Token 설정 (선택 사항이지만 권장)
# os.environ["HF_TOKEN"] = "YOUR_HF_TOKEN_HERE"

IMG_MODEL_NAME = 'clip-VIT-B-32' # vision transformer : 멀티모달 가능 모델.
TEXT_MODEL_NAME = "sentence-transformers/clip-VIT-B-32-multilingual-v1" # clip 확장 버전, 한국어 가능
img_model = SentenceTransformer(IMG_MODEL_NAME) # 이미지 Embedding용 모델
text_model = SentenceTransformer(TEXT_MODEL_NAME) # 텍스트 Embedding용 모델

cilent = PersistentClient(path=CHROMA_DIR)
col = cilent.get_or_create_collection(COLLECTION)
```

```python
from numpy import delete
# 이미지 경로를 수집하는 수집용 유틸 함수 생성
def list_imagesFunc(root:str, exts=('.jpg','.jpeg','.png','.gif','.webp')) -> List[str]:
  return [str(p) for p in Path(root).rglob("*") if p.suffix.lower() in exts]
#rglob : root폴더속을 재귀적으로 뒤진다.
# -> 확장자가 exts의 값에 맞는파일만 받아서 List를 만듦

# VectorDB 구축하기 : 이미지들을 CLIP Embedding로 저장
def build_image_dbFunc():
  Path(IMG_DIR).mkdir(parents=True, exist_ok=True)
  paths = list_imagesFunc(IMG_DIR)
  # print(paths)
  if not paths:
    print("해당 이미지 파일이 없음.")

  # VectorDB를 새로 구축할 때 기존 레코드 전체 삭제 후 새로 구축 - 실습용
  existing  = col.get(include=[]) # include를 비우면 ids만 반환
  ids_to_delete = existing.get("ids", []) # ids가 없으면 빈 리스트 반환
  if ids_to_delete:
    col.delete(ids=ids_to_delete) # 해당 ids 모두 삭제
    print(f'기존 {len(ids_to_delete)}개 레코드 삭제 완료')

  print(f"{len(paths)}개 이미지 로딩")

  # PIL.Image 리스트를 이미지용 모델로 Embedding
  imgs = [Image.open(p).convert('RGB') for p in paths]
  vecs = img_model.encode(imgs, normalize_embeddings=True).tolist()
  ids = [f"img_{i}" for i in range(len(paths))]
  metadatas = [{"path":p} for p in paths]
  col.add(
      ids=ids,
      embeddings=vecs,
      metadatas=metadatas
  )
```

```python
# Text로 이미지 검색(VectorDB에서 검색) 함수 생성
def retrieve_imageFunc(query:str, top_k:int=3) -> Dict[str, Any]:
  q_vec = text_model.encode([query], normalize_embeddings=True).tolist()[0]
  # print(q_vec)
  print('컬렉션 자료 수', col.count())
  res = col.query(
      query_embeddings=[q_vec],
      n_results=top_k,
      include=["metadatas", "distances"],
  )
  metas = res['metadatas'][0] if res['metadatas'] else []
  dists = res['distances'][0] if res['distances'] else []
  # paths = [m['path'] for m in metas] # 같은 의미이고 python은 가독성을 위해 타입을 줄 수 있다.
  paths:List[str] = [m.get('path', '?') for m in metas]

  return {
      "query":query,      # 원래 질의 문장
      "image_paths":paths, # 질문에 대한 추천된 이미지 경로 top_k개
      "distences":dists,  # 질문에 대한 cos거리값 top_k개
  }
# build_image_dbFunc()
# retrieve_imageFunc("귀여운 것들")

# 텍스트 가시성 좋게 출력하는 함수 생성
def debug_print_retrieveFucn(result:Dict[str, Any]):
  print("이미지 검색 결과")
  print("질의 : ", result['query'])
  print("검색 결과")
  for i ,(p, d) in enumerate(zip(result['image_paths'], result['distences'])):
    print(f"    {i}){p} (distence={d:.4f})")

# 검색 결과 이미지 출력
def show_images(result:Dict[str, Any], max_cols:int=3):
  paths = result['image_paths']
  if not paths:
    print('표시할 이미지가 없음.')
    return
  for i, p in enumerate(paths, start=1):
    try:
      img = Image.open(p).convert("RGB")
      print(f"{i}, {p}")
      display(img) # colab용 이미지 출력 방법

    except Exception as e:
      print(f"{i}, {p} 이미지 로딩 실패 : {e}")

imsi = retrieve_imageFunc("귀여운 강아지, 고양이")
debug_print_retrieveFucn(imsi)
show_images(imsi)
```

```python
# LangChain 구성하기
parser = StrOutputParser()

# LLM에게 추천된 이미지 설명용 자료 얻기
def encode_image_to_data_urlFunc(img_path:str) -> str:
  # jpg -> jpeg로 변환하기를 LangChain이 원함
  ext = Path(img_path).suffix.lower().replace('.','') # '.'을 기준으로 확장자만 추출
  if ext == 'jpg':
    ext = 'jpeg'
  if not ext:
    ext = 'jpeg'

  supported_exts = {'jpeg','png','gif','webp'} # OpenAI 입력에서 주로 사용하는 이미지 mime type
  if ext not in supported_exts:
    raise ValueError(f"지원하지 않는 이미지 형식 : {ext}")

  with open(img_path, "rb") as f:
    b64 = base64.b64encode(f.read()).decode('utf-8')

  return f"data:image/{ext};base64,{b64}" # url만들어서 보냄.

# print(encode_image_to_data_urlFunc("images_db/ani1.jpeg"))

# LLM에게 Text +여러 이미지가 포함된 HumanMessage를 생성
def build_multimodal_messageFunc(query:str):
  res = retrieve_imageFunc(query, top_k=3)

  contents = [
      {
          'type':'text',
          'text':(
              "사용자 설명 : " + res['query'] + "\n"
              "아래에 표시된 여러 사진을 보고 장면과 분위기를 한국어로 자세히 설명해 줘\n"
              "사진 순서대로 1번 제목, 2번 제목, 3번 제목 ... 제목을 이미지를 보고 생성하고 번호를 붙여서 설명해\n"
              "마크다운은 절대로 표시하지마"
          ),
      }
  ]

  for idx, path in enumerate(res['image_paths'], 1):
    try:
      data_url = encode_image_to_data_urlFunc(path)
      contents.append({
          'type':'text',
          'text':f"{idx}번 이미지 (파일 경로 : {path})",
      })
      contents.append({
          'type':'image_url',
          'image_url':{'url':data_url},
      })

    except Exception as e:
      # 이미지 인코딩 실패시 로그 텍스트만 출력하기
      contents.append({
          "type":"text",
          'text':f"{idx}번 이미지 로딩 실패 : {path}, 오류 원인 : {e}"
      })
  return [HumanMessage(content=contents)]

# LCEL 구성하기
image_search_chain = (
    RunnableLambda(build_multimodal_messageFunc)
    | llm
    | parser
)

# 실행 함수 생성
def runFunc():
  print("1) 이미지 구축 DB")
  build_image_dbFunc()

  print("2) Text로 VectorDB의 유사 이미지 검색")
  qestion = "사랑스럽고 귀여운 아기 동물들 이미지 찾아 줘"
  result = retrieve_imageFunc(qestion, top_k=3)
  print("result : ", result)

  debug_print_retrieveFucn(result=result)
  show_images(result=result)

  print("3) LangChain으로 유사 이미지 검색 결과를 LLM이 설명하는 곳")
  expl = image_search_chain.invoke(qestion)
  print("[추천 설명 결과]\n")
  print(expl)

if __name__=="__main__":
  runFunc()
```
