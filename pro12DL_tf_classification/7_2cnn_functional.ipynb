```python
# MNIST dataset으로 CNN 실습 - 함수형 API
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# 1) 데이터 준비
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
# print(x_train[0])
print(x_train.shape) # (60000, 28, 28)

# 채널(channel) 추가 (흑백인 경우 1) -> 구조 변경후(3차원 -> 4차원 : CNN이 필요한 차원으로 변경) 정규화
x_train = x_train.reshape(-1, 28, 28, 1).astype('float32') / 255.0
x_test = x_test.reshape(-1, 28, 28, 1).astype('float32') / 255.0
print(x_train.shape) # (60000, 28, 28, 1)
```

```python
"""
# 2) 모델 정의 1
inputs = tf.keras.layers.Input(shape=(28, 28, 1))

# Convolution Block1 생성 ------------------------------- CNN은 Convolution, Pooling이 끝이야! 이해하는게 중요
x = tf.keras.layers.Conv2D(filters=16,kernel_size=(3,3), padding='same', activation='relu')(inputs)
# Pooling
x = tf.keras.layers.MaxPool2D(pool_size=(2,2))(x)
x = tf.keras.layers.Dropout(rate=0.2)(x) # 옵션

# Convolution Block2 생성
x = tf.keras.layers.Conv2D(filters=32, kernel_size=(3,3), padding='same', activation='relu')(x)
x = tf.keras.layers.MaxPool2D(pool_size=(2,2))(x)
x = tf.keras.layers.Dropout(rate=0.2)(x) # 옵션

# FC layer(Fully Connected Layer) - 1차원으로 구조 변경
x = tf.keras.layers.Flatten()(x)

# Dense
x = tf.keras.layers.Dense(units=64, activation='relu')(x)
x = tf.keras.layers.Dropout(rate=0.2)(x)
x = tf.keras.layers.Dense(units=32, activation='relu')(x)
x = tf.keras.layers.Dropout(rate=0.2)(x)
outputs = tf.keras.layers.Dense(units=10, activation='softmax')(x)

model = tf.keras.models.Model(inputs=inputs, outputs=outputs, name='mnist_cnn_func')
print(model.summary()) # 한 셀안에 print가 하나만 있으면 print 안적어도 됨 - 쓰는게 좋다.
"""
```

```python
# 2) 모델 정의 2
inputs = tf.keras.layers.Input(shape=(28, 28, 1))

# Convolution Block1 생성 -------------------------------
x = tf.keras.layers.Conv2D(filters=32,kernel_size=(3,3), padding='same', use_bias=False)(inputs)
x = tf.keras.layers.BatchNormalization()(x)
# x = tf.keras.layers.Activation('relu')(x)
x = tf.keras.layers.ReLU()(x)
# use_bias=False : ReLu 레이어를 분리 -> BatchNormalization 뒤에 ReLu를 둠.
# BatchNormalization() : 과적합 방지 - 학습 안정화, 수렴(weight값) 가속화 사용할때 ReLu를 뒤에 두는게 좋다.

# Pooling
x = tf.keras.layers.MaxPool2D(pool_size=(2,2))(x)
x = tf.keras.layers.Dropout(rate=0.2)(x) # 옵션

# Convolution Block2 생성 -------------------------------
x = tf.keras.layers.Conv2D(filters=64,kernel_size=(3,3), padding='same', use_bias=False)(inputs)
x = tf.keras.layers.BatchNormalization()(x)
# x = tf.keras.layers.Activation('relu')(x)
x = tf.keras.layers.ReLU()(x)

# Pooling
x = tf.keras.layers.MaxPool2D(pool_size=(2,2))(x)
x = tf.keras.layers.Dropout(rate=0.2)(x) # 옵션

# FC layer(Fully Connected Layer) - 1차원으로 구조 변경
x = tf.keras.layers.Flatten()(x)

# Dense
x = tf.keras.layers.Dense(units=64, activation='relu')(x)
x = tf.keras.layers.Dropout(rate=0.2)(x)
x = tf.keras.layers.Dense(units=32, activation='relu')(x)
x = tf.keras.layers.Dropout(rate=0.2)(x)
outputs = tf.keras.layers.Dense(units=10, activation='softmax')(x)

model = tf.keras.models.Model(inputs=inputs, outputs=outputs, name='mnist_cnn_func')
print(model.summary())
```

```python
# 3) compile
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

es = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
hist = model.fit(
    x_train, y_train, epochs=100, batch_size=128, validation_split=0.1, callbacks=[es], verbose=1
)
```

```python
# 4) 모델 평가하기 (train, test평가 점수의 차이가 크면 overfitting 의심)
train_loss, train_acc = model.evaluate(x_train, y_train, verbose=0)
print(f'train loss : {train_loss:.4f}, train_acc : {train_acc:.4f}')

test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)
print(f'test loss : {test_loss:.4f}, test_acc : {test_acc:.4f}')
```
