# 번역 Scaled Dot-Product Attention (Tokenizer + Attention)

```python
import numpy as np
import math
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
# Encoding

# 입출력 문장
input_text = 'I have a pen'
output_text = '<sos> 나는 펜을 갖고 있다 <eos>'

# 토큰 처리
input_tokenizer = Tokenizer(
    filters='' , # 특수문자 제거 X
    lower = False, # 소문자로 안바꿈
)
output_tokenizer = Tokenizer(
    filters='' , # 특수문자 제거 X
    lower = False, # 소문자로 안바꿈
)

# 단어 사전 생성
input_tokenizer.fit_on_texts([input_text])
output_tokenizer.fit_on_texts([output_text])
print(input_tokenizer)
print(output_tokenizer)
```
