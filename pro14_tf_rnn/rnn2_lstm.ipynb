# LSTM으로 다음 숫자 예측

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, GRU, Input

x = np.array([[1, 2, 3],[2, 3, 4],[3, 4, 5],[4, 5, 6],[5, 6, 7],[6, 7, 8],[7, 8, 9],[8, 9, 10]])
y = np.array([4, 5, 6, 7, 8, 9, 10, 11])

x = x.reshape((8, 3, 1)) # 입력형태 : (samples, time_steps, features)

model = Sequential()
model.add(Input(shape=(3, 1))) # 시계열(time_steps) 3개 , features 1개
model.add(LSTM(
    units=32,
    activation='tanh', # default값
))
model.add(Dense(16, activation='relu'))
'''
model.add(LSTM( # 표준
    units=10,activation='tanh', # default값
    return_sequences=True, # default값 - many2many
))
model.add(GRU( # LSTM에 비해 계산량이 적고 속도는 빠름
    units=10,activation='tanh', # default값
))
'''
model.add(Dense(units=1, activation='linear'))
model.summary()
```

```python
model.compile(loss='mse', optimizer='adam')

from tensorflow.keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(monitor='loss', patience=3, mode='auto')
model.fit(x, y, epochs=2000, batch_size=1, verbose=2, callbacks=[early_stopping])
```

```python
print('예측값 :', model.predict(x).ravel())
print('실제값 :', y.ravel())

# 새로운 값 예측하기
x_input = np.array([0, 1, 2])
x_input = x_input.reshape((1, 3, 1))
print('예측값 :', model.predict(x_input))

new_pred = model.predict(x_input)
print('새로운 값 예측 : ', new_pred.ravel())
```
