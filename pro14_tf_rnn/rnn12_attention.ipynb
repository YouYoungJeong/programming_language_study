# Attention
## 개념 구조 이해
- Query - Key - Value 구조를 이용한 Dot Product Attention 계산하기
- Qurey와 Key간의 유사도를 내적으로 계산하고, 그 결과로 Value를 가중합 하여 출력 벡터 만들기
- Query(디코더의 입력 단어), Key(디코더 관심 후보, 인코더에 있는 값들), Value(key와 제일비슷한 데이터값)

```python
# 간단한 숫자 예제로 Attention이 어디에 집중할 지를 학습하고, 그 정보를 바탕으로 출력값을 가중합하기

import numpy as np

Q = np.array([2.0, 1.0]) # 내가 현재 집중해 보고싶은 벡터(디코딩된 벡터중 하나), 현재 상태(decoder hidden state)
# 예를 들어 두가지 값이 있는데 [달콤함, 새콤함] 라고 가정

K = np.array([ # 입력 단계(인코더)에서 나온 각 토큰의 특징 벡터들
    [1.0, 0.0],
    [0.0, 1.0],
    [1.0, 1.0],
])

V = K.copy()  # 입력 토큰이 실제로 가진 정보(출력할 때 가중합할 벡터)

scores = K.dot(Q) # Q와 K를 비교해 내적을 구함 - 점수계산 [1*2 + 0*1, 0*2 + 1*1, 1*2 + 1*1] = [2.0, 1.0, 3.0]

# softmax로 Attention Weight(가중치) 구하기
def softmaxFunc(x):
  e = np.exp(x - np.max(x)) # - : x값이 너무 크면 값이 안나와
  return e / e.sum()

weights = softmaxFunc(scores) # 어떤 단어를 얼마나 볼지 비율료 바꿈
print(weights)
print(scores)

# Values에 가중치를 곱해 합치면 최종 Attention출력

output = (weights[:, None] * V).sum(axis=0)
print('Attetion 최종 출력값',output) # <= Attention의 최종 출력
# 핵심은 "중요한 단어의 Value는 크게 반영 하고, 덜 중요한 단어는 조금만 반영해서 하나의 결과를 만든다"

# 최종 출력 계산은 이런 식
# output_imsi = 0.24472847*[1.0, 0.0] + 0.09003057*[0.0, 1.0] + 0.66524096*[1.0, 1.0]
out1 = 0.24472847*1 + 0.09003057*0 + 0.66524096*1
out2 = 0.24472847*0 + 0.09003057*1 + 0.66524096*1
print(out1, out2) # [달콤함, 새콤함]기준으로 보면 달콤함=0.909,새콤함=0.755 정도가 섞인 최종 정보

print('Score:',np.round(scores, 3))
print('Weights:',np.round(weights, 3))
print('Output:',np.round(output,3))
```
