# RAG, LangChain, LangGraph 간단 정리

## RAG(Retrieval-Augmented Generation)

* 사용자의 질문과 관련된 자료를 먼저 검색한 뒤, 그 내용을 바탕으로 답변을 생성하는 방식
* 모델이 기억하고 있는 지식만 사용하는 것이 아니라, 외부 문서나 데이터베이스를 참고해서 답변함
* 예: 회사 문서, PDF, 논문, DB를 검색해서 답변하는 챗봇

## LangChain

* RAG 같은 AI 서비스 흐름을 쉽게 만들 수 있게 도와주는 프레임워크
* LLM, 검색기, 벡터DB, 프롬프트, 도구 사용 등을 연결해주는 역할
* 예: “문서 검색 → 관련 내용 추출 → GPT에게 전달 → 답변 생성” 과정을 코드로 쉽게 구성

## LangGraph

* LangChain 기반에서 작업 흐름을 그래프 구조로 더 정교하게 설계할 수 있게 해주는 도구
* 노드(Node)와 엣지(Edge)를 이용해 AI가 어떤 순서로 작업할지 정의함
* 단순한 RAG보다 복잡한 조건 분기, 반복, 에이전트 흐름을 만들 때 사용
* 예: “검색 → 답변 생성 → 검증 → 부족하면 다시 검색 → 최종 답변” 같은 구조

## Vector DB

* 문서나 문장을 숫자 벡터 형태로 저장해두는 데이터베이스
* 사용자의 질문과 의미가 비슷한 문서를 빠르게 찾기 위해 사용
* 예: Chroma, FAISS, Pinecone

## Embedding

* 문장이나 문서를 숫자 벡터로 변환하는 과정
* 의미가 비슷한 문장끼리 가까운 벡터가 되도록 만들어 검색에 활용함
* 예: “탱크 탐지 모델 성능”이라는 질문과 관련된 문서를 찾기 위해 사용

## Retriever

* 사용자의 질문과 관련 있는 문서를 찾아오는 검색기
* 보통 Vector DB에서 의미가 비슷한 문서를 검색하는 역할을 함

## Prompt

* LLM에게 어떤 방식으로 답변해야 하는지 알려주는 입력 문장
* 질문, 조건, 역할, 답변 형식 등을 포함할 수 있음
* 예: “아래 문서를 참고해서 한국어로 요약해줘”

## Agent

* LLM이 단순히 답변만 하는 것이 아니라, 필요한 도구를 선택해서 작업을 수행하는 구조
* 검색, 계산, 파일 읽기, API 호출 등을 상황에 따라 사용할 수 있음

## Node

* LangGraph에서 하나의 작업 단위를 의미함
* 예: 문서 검색 노드, 답변 생성 노드, 검증 노드

## Edge

* LangGraph에서 노드와 노드를 연결하는 흐름을 의미함
* 어떤 작업 다음에 어떤 작업을 실행할지 정의함

## 전체 흐름 예시


사용자 질문 입력
  → 질문을 Embedding으로 변환
  → Vector DB에서 관련 문서 검색
  → Retriever가 관련 문서 가져오기
  → Prompt에 질문과 문서 내용 넣기
  → LLM이 답변 생성
  → 필요하면 LangGraph로 검증 또는 재검색 진행

## HF Hub(Hugging Face Hub)

* AI 모델, 데이터셋, 학습 결과물, 데모 앱을 올리고 공유할 수 있는 플랫폼
* GitHub가 코드 저장소라면, HF Hub는 AI 모델과 데이터셋 저장소에 가까움
* 공개된 모델을 다운로드해서 바로 사용할 수 있고, 내가 학습한 모델도 업로드할 수 있음
* 예: YOLO 모델, LLM 모델, 이미지 분류 모델, 의료 영상 모델, 텍스트 임베딩 모델 등

## 주로 저장되는 것

* Model: 학습된 AI 모델 가중치
* Dataset: 학습/검증용 데이터셋
* Spaces: 웹에서 실행 가능한 AI 데모 앱
* Tokenizer: LLM 입력 문장을 토큰으로 바꾸는 도구
* Config: 모델 구조와 설정 파일

## 예시

* `bert-base-uncased`: 자연어 처리 모델
* `sentence-transformers/all-MiniLM-L6-v2`: 문장 임베딩 모델
* `meta-llama/...`: LLaMA 계열 LLM 모델
* `openai/clip-vit-base-patch32`: 이미지-텍스트 임베딩 모델

## RAG에서 HF Hub 역할

* Embedding 모델을 HF Hub에서 가져와 사용 가능
* LLM 모델을 HF Hub에서 다운로드해서 로컬에서 실행 가능
* 문서 검색용 모델이나 reranker 모델도 HF Hub에서 가져올 수 있음

# Knowledge Data와 Vector Data 정리

## Knowledge Data

* AI가 답변할 때 참고하는 원본 지식 자료
* 사람이 읽을 수 있는 형태의 데이터
* RAG에서 검색 대상이 되는 기본 자료
* 예: PDF, 문서, 회의록, 논문, CSV, 매뉴얼, 프로젝트 보고서, 데이터베이스 내용

## Vector Data

* Knowledge Data를 AI가 검색하기 쉽게 숫자 벡터로 변환한 데이터
* 문서나 문장의 의미를 숫자 배열로 표현한 형태
* 사람이 직접 읽기보다는 검색 시스템이 사용하는 데이터
* Vector DB에 저장되어 의미 기반 검색에 사용됨

## 차이점

| 구분     | Knowledge Data            | Vector Data                |
| ------ | ------------------------- | -------------------------- |
| 의미     | 원본 지식 자료                  | 원본 자료를 숫자로 변환한 데이터         |
| 형태     | 문서, 텍스트, PDF, CSV 등       | 숫자 벡터                      |
| 사람이 읽기 | 가능                        | 어려움                        |
| 사용 목적  | AI가 참고할 원본 정보             | 관련 문서를 빠르게 검색              |
| 저장 위치  | 파일 저장소, DB, 문서 폴더 등       | Vector DB                  |
| 예시     | “YOLO 모델 mAP50은 0.8575이다” | `[0.12, -0.45, 0.88, ...]` |

## RAG에서의 흐름

1. PDF, 문서, CSV 같은 Knowledge Data를 준비함
2. 문서를 작은 단위로 나눔
3. 각 문장을 Embedding 모델로 숫자 벡터로 변환함
4. 변환된 Vector Data를 Vector DB에 저장함
5. 사용자가 질문하면 질문도 벡터로 변환함
6. 질문 벡터와 비슷한 Vector Data를 검색함
7. 검색된 원본 Knowledge Data를 LLM에게 전달함
8. LLM이 그 내용을 바탕으로 답변을 생성함

```python
from openai import OpenAI
from sentence_transformers import SentenceTransformer
import chromadb
from chromadb.config import Settings
import os
from dotenv import load_dotenv

load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
model = "gpt-4o-mini"
embedder = SentenceTransformer('all-MiniLM-L6-v2') # sentence-transformers/all-MiniLM-L6-v2
```

# RAG 첫단계 : Text를 읽어 VectorDB에 저장 (데이터 수집)

```python
from langchain_core import documents
from pydantic_core.core_schema import DataclassSchema
# 방법 1 - 문서데이터 직접 작성
# documents = [
#     "김치찌개는 한국의 대표적인 찌개 요리이다",
#     "된장찌개는 발효된 된장을 이용해 만든다",
#     "비빔밥은 여러 가지 나물을 비벼 먹는 밥 요리이다",
#     "불고기는 양념한 소고기를 구워 먹는 전통 음식이다",
#     "삼계탕은 닭에 인삼, 대추 등을 넣고 푹 끓인 보양식이다."
# ]

# 방법 2: 문서 파일 읽기 (Python사용)
# with open('foods.txt', 'r', encoding='utf-8') as f:
#   documents = [line.strip() for line in f if line.strip()]
# documents = [doc.split() for doc in documents]
# print(documents)

# 방법 3: 문서 파일 읽기 (LangChain 사용)
from langchain_community.document_loaders import TextLoader # 파일 읽을때 사용

# loader = TextLoader('foods.txt')
# datas = loader.load()
# documents = datas[0].page_content.split('\n')
# # documents리스트에서 빈 문자열 제거 후 , 앞뒤 공백을 잘라 정리
# documents = [doc.split() for doc in documents if doc.strip()]
# print(documents)

# 방법 4: 문서 파일 읽기 (LangChain)
from langchain_text_splitters import CharacterTextSplitter # 텍스트를 분할할때 사용

loader = TextLoader('foods.txt')
datas = loader.load()
# text 나누기  - chunck단위로 쪼갠다
splitter = CharacterTextSplitter(separator='\n', chunk_size=100, chunk_overlap=0)
spl_docs = splitter.split_documents(datas) # Document 객체를 여러 조각으로 분할

# Document 객체에서 실제 문자열만 추출
documents = [doc.page_content for doc in spl_docs]
print(documents)
print(len(documents))
print(documents[0])
```

# 임베딩 벡터로 변환 후 DB에 저장

```python
doc_embeddings = embedder.encode(documents)
print(doc_embeddings[0][:3])

# ChromaDB에 저장
# 방법 1 : 임시 저장(Ram에다가만 저장, 디렉토리를 적지 않음.)
# chroma_client = chromadb.Client(Settings(anonymized_telemetry=False)) # anonymized_telemetry : 보안을 목적으로,익명으로 수집하지마

# 방법 2 : 반영구 저장
# chroma_client = chromadb.PersistentClient(path="./chroma_db") # 세부 설정 불가
chroma_client = chromadb.Client(
    Settings( # 세부 설정 가능
        persist_directory="./chroma_db",
        anonymized_telemetry=False # 익명으로 자료 수집 불가 (보안 강화 목적)
    )
  )
# 기존 데이터 지우고(delete) collection 새로 작성(create)
# chroma_client.delete_collection("foods")
# collection = chroma_client.create_collection(name="foods")

collection = chroma_client.get_or_create_collection(name="foods")

# ChromaDB에 저장
for i, (doc, embedding) in enumerate(zip(documents, doc_embeddings)):
  collection.add(
    documents = [doc],
    embeddings = [embedding.tolist()],
    ids = [f"doc_{i}"]
  )
```

# 다음 단계 : 질문 -> 관련문서 -> LLM 응답 생성

```python
# RAG(Retrieval 단계) : 관련문서 검색
# 사용자의 기본 질문 - 이렇게만 하면 결과가 좋지 않아서 여기저기에서 cos유사도를 이용해 가져옴(질문 강화)
query = '한국의 대표적인 찌개 음식을 알려줘'

# 질문과 유사한 문자열을 VeatorDB에서 읽기 - 질문도 Vector화 시켜야함
query_embedding = embedder.encode([query], normalize_embeddings=True)[0] # 정규화된 embedding Vector로 변환
print(query_embedding[:2])

results = collection.query(
  query_embeddings = [query_embedding.tolist()],
  n_results = len(documents),
  include = ["documents","distances"]
)
print("results : \n", results)

# 질문에 대한 거리값 목록 출력
import numpy as np
distences = results['distances'][0]
print(f"거리(distence)값 : {distences}")

# 직접 계산한 유사도 값으로 확인
similarities = []
for doc_id in results['ids'][0]:
  doc_embed = collection.get(ids=[doc_id], include=['embeddings'])['embeddings'][0]
  doc_embed = np.array(doc_embed, dtype=float) # 계산을 위해 numpy 배열로 변환
  sim = np.dot(query_embedding, doc_embed)
  similarities.append(sim)
print(f"직접 계산한 유사도(similarity) 값 : {similarities}")

ids_list = results["ids"][0]
docs_list = results["documents"][0]
dist_list = results["distances"][0]

for i, doc_id in enumerate(ids_list):
  print("===="*30)
  print(f"문서 {i}")
  print(f"  - id:{doc_id}")
  print(f"  - 문서내용:\n{docs_list[i]}")
  print(f"  - 거리(distence):{dist_list[i]:.4f}")
  print(f"  - 유사도(similarity):{similarities[i]:.4f}")
```

```python
# RAG(generation 단계) : 검색된 문서로 질문을 보강해서 LLM에서 답변 요청 후 생성
retrieved_docs_list = results['documents'][0] # ChromaDB에서 검색된 문서 list 꺼내기
retrieved_docs = "\n\n".join(retrieved_docs_list)
print(retrieved_docs)

prompt = f"""
  너는 한국 전통 음식 전문가야
  지금부터 사용자 질문에 답변 할 때는 반드시 아래 문서 내용을 참고해.

  [참고문서]
  {retrieved_docs}

  [사용자 질문]
  {query}

  위 참고 문서를 바탕으로 10문장 이내로 답해 줘
  마크다운 기호 없이 평문으로 답해 줘
"""
print(prompt)
```

```python
# LLM 응답 요청 - OpenAI 서버에게 요청
response = client.responses.create(
    model = model,
    input = prompt
)

import re

text = response.output_text
# print(text)
sentences = re.split(r'(?<=[.!? ])\s+', text) # 마침표, 물음표, 느낌효 뒤에서 문장 분리
for s in sentences:
  s = s.strip()
  if not s:
    continue

  print(s)
```
